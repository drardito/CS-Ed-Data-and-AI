## Data Analysis/Visualization in R

For this work, we will explore and play with the first two sections of the book, *Whole Game* and *Visualize*. The section *Whole Game* is a survey of the basic kinds of data analysis tasks and functions most commonly used for this kind of work, such as importing, transforming, and tidying data for analysis. The second *Visualize* (surprise, surprise) is a survey of basic kinds of methods of visualizing data. I think you will find both will build nicely on the R skills and techniques you have already explored.
