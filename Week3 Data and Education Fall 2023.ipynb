[Digital Humanities](https://en.wikipedia.org/wiki/Digital_humanities) is a field of study that applies computational tools to the study of the humanities. Last week, we took at peak at this approach by using both unplugged tools (like highlighters and counting) and plugged tools (Voyant Tools) to do a small investigation into the first paragraph of Moby Dick. The goal of all this was to wade into the world of Digital Humanities and to treat text as data. This week and next, we will be expanding our toolbox in order to take a much deeper dive into the use of computational tools to explore, investigate, analyze, and visualize text.

This work will start by using a tool called Plotting Plots and then move to text analysis functionality in the R programming Language.




## Work for Week 3 - Computational Tools for Textual Analysis



### Chunk 1: Plotting Plots

<img src="https://i.imgur.com/hX31W3f.png" width="650" height="340">

A friend and colleague, Tom Liam Lynch, is a former high school English teacher and teacher educator, who has focused on making computationally supported textual analysis available to English teachers. Part of his work has included the development of [Plotting Plots](https://plottingplots.com/), a website where teachers and students can explore texts computationally.

Here is the work for this Chunk:

1.  *Reading*. Read Tom's guide to Plotting Plots, [For Book Lovers for Like Data](https://plottingplots.com/wp-content/uploads/2021/09/plotting-plots-classroom-DRAFT-2.pdf) in which Tom describes his work and ways that teacher can use this work with their students. By the way, Tom's work is focused on English teachers and their students, but all of our students read texts of various kinds, and these types of tools can be used with all kinds of texts (more later).
2. *Guided Practice*. Now that you have a sense of what Tom is up to and what we mean by this type of analysis, it's time  for you to start using Plottig Plot. I ask that you start by reading two or three of [Tom's posts](https://plottingplots.com/posts/), and I strongly encourage you to start with his analysis of [spells in the Harry Potter series](https://plottingplots.com/top-spells-in-harry-potter-series-by-book/). As you read these posts try out what he is discussing as you read along. 
3. *Exploring on Your Own*. Now that you have some familiarity with this tools, go to [Make Some Plots](https://plottingplots.com/plots/), choose a book or two (or three), and plot away. Please collect artifacts as you go (notes, screenshots, etc.), so that you can reflect on your explorations of which books you choose. Here are some prompts to get started: which books did you choose and why? what did your plots reveal about the book? what further questions did these analyses raise? 


### Chunk 2: Text Analysis in R

<img src="https://uc-r.github.io/public/images/analytics/descriptives/tidy_text1.png" width="650" height="340">

Now that we have used some unplugged tools for data analysis and explored some computational tools (Plotting Plots and Voyant-Tools), it's time to explore some coding tools that support the kinds of textual analysis and visualization in which we have been engaged. This type of work can be done productively in both Python and R, but for this week, we will focus on a set of R tools.
Here is the work for this chunk:

1. *Getting Ready*. in the [overview to this course](https://github.com/drardito/CS-Ed-Data-and-AI/blob/main/Data%20and%20Education%20Fall%202023.md), I asked that you download R and RStudio to your computer. If you haven't already done so, please do so now. Once you are set up in RStudio, you will need to install the packages that support this type of text analysis and visualization. should install these packages: *tidytext*, *dplyr*, *ggplot2*, *tidyr*, and *broom*. You can do this by going to Tools>Install Packages in RStudio.

2. *Get Set*. There is an excellent and freely available book written by the developers of these tidytext tools in R, [Text Mining with R](https://www.tidytextmining.com/) by Julia Silge and David Robinson. Note that the online version of the book also contains links to GitHub repository which has the text, images, and code used in each chapter, which I have found very useful. Please start by reading the book's Preface.

3. *Go*. Work through two chapters: Chapter 1 The Tidy Text Format and Chapter 3 Analyzing Word and Document Frequency: tf-idf. By work through, I mean read them and follow the examples, and do perform each exercise in RStudio. Please save the code and visualizations you generate so you can share them.

4. *Going Further*. Now that you have some experience using R to analyze texts and visualize those analyses, it times to go beyond the exercises. Use a text of your choosing and do similar word frequency analyses to and related visualizations to show what you now know.
